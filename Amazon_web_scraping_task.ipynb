```python
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
```

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://www.amazon.eg/s?k=samsung&rh=p_89%3Asamsung&language=en&ref=SQEG-WEB-SR301")
```

```python
Title=[]
Image=[]
Price = []
Total_Rate=[]
Rate=[]
```

```python
all_product=driver.find_elements('xpath',"//div[@data-component-type='s-search-result']")
```

```python
for i in all_product:
    try:
        title=i.find_element('xpath',".//h2//a//span").text
    except:
        title=""
    
    try:
        img=i.find_element('xpath',".//img[@class='s-image']").get_attribute('src')
    except:
        img=""
    
    try:
         p=i.find_element(By.CSS_SELECTOR, '.a-price-whole').text
    except:
         p=""
    
    try:
         total_rate=i.find_element('xpath',".//span[@class='a-price-whole']").text
    except:
         total_rate=""
    
    try:
         rating = i.find_element(By.CSS_SELECTOR, '.a-icon-alt').get_attribute('innerHTML')
    except:
         rating=""
    
    
    Title.append(title)
    Image.append(img)
    Price.append(p)
    Total_Rate.append(total_rate)
    Rate.append(rating)
```

```python
Data = {
    'Title': Title,
    'Price': Price,
    'Image': Image,
    'Rate': Rate,
    'Total_Rate': Total_Rate
}
df = pd.DataFrame(Data)

df
```

```python
df.to_csv('all_products.csv')
```
